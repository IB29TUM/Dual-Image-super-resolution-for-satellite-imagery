```python
import os
import shutil

def make_copies_of_images(folder_path, num_copies=5):
    if not os.path.exists(folder_path):
        print(f"Error: Folder '{folder_path}' does not exist.")
        return

    for root, dirs, files in os.walk(folder_path):
        for filename in files:
            if not filename.lower().endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp')):
                continue

            image_path = os.path.join(root, filename)
            for i in range(num_copies):
                new_filename = f"{os.path.splitext(filename)[0]}_copy_{i+1}{os.path.splitext(filename)[1]}"
                new_image_path = os.path.join(root, new_filename)
                shutil.copy(image_path, new_image_path)

if __name__ == "__main__":
    folder_path = "hr_images/"  # Replace this with the path to your hr_image folder
    num_copies = 4
    make_copies_of_images(folder_path, num_copies)
    print(f"Copied {num_copies} copies of each image in the '{folder_path}' folder.")
```
